```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, tight_layout = True)

# linear
axes[0,0].plot([1, 2, -5, 2])

# sin
x = np.linspace(0, 10, 100)
y = np.sin(2 * x)
axes[0,1].plot(x,y)

# hist
x = np.random.standard_normal(1000)
axes[0,2].hist(x)

# scatter
x = np.random.normal(0, 2, 60)
y = np.random.normal(0, 2, len(x))
axes[1,0].scatter(x, y)
plt.show()
```
