## データ準備

```python
from sklearn.datasets import load_wine
wine_dataset = load_wine()
wine_dataset
```

```python
wine_dataset.feature_names
```

```python
import pandas as pd
df = pd.DataFrame(data=wine_dataset.data, columns = wine_dataset.feature_names)
df.head()
```

```python
wine_dataset.target
```

```python
df['category'] = wine_dataset.target
df.head()
```

```python
df.shape
```

```python
# データを目的変数と説明変数に分割
X = wine_dataset.data
y = wine_dataset.target
```

```python
# さらに各変数を学習用とテスト用に分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## モデル学習

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=5)

# 課題用 ランダムフォレストモデル
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=5)
```

```python
model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
```

## 性能評価

```python
dt_pred = model.predict(X_test)
rf_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"accuracy score of decision tree :{accuracy_score(y_test, dt_pred)}")
print(f"accuracy score of random forest :{accuracy_score(y_test, rf_pred)}")
```
